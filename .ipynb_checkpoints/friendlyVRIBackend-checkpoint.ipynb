# Load the Observation Managaer

```python
%matplotlib inline

from Imports.vriCalc import observationManager
import numpy as np
import matplotlib as mpl

import matplotlib.pyplot as plt
from matplotlib.figure import Figure
from matplotlib.colors import LogNorm
from matplotlib.ticker import MaxNLocator
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.backends.backend_tkagg import NavigationToolbar2Tk

obsMan = observationManager(verbose=True, debug=True)
obsMan.get_available_arrays()
```

# Select Array Configurations and Hour-Angle Ranges

```python
obsMan.select_array('VLA_A')
obsMan.select_array('VLA_B')
obsMan.select_array('VLA_C')
obsMan.select_array('VLA_D')
obsMan.get_selected_arrays()
```

# Set the Observing Frequency (MHz) and Source Declination (deg)

```python
obsMan.set_obs_parms(1420.0, 20.0)
```

# Calculate the UV-Coverage

```python
obsMan.calc_uvcoverage()
```

## To Plot UV-Covergae

```python
uvCovArr = obsMan.arrsSelected
```

```python
colLst=["r", "b", "g", "m", "c", "y", "k"]
if uvCovArr is not None:
    oLst = []
    sLst = []
    for i, e in enumerate(uvCovArr):
        oLst.append(i)
        sLst.append(e["scaleMin_deg"])
    multiLst = list(zip(sLst, oLst))
    multiLst.sort()
    
    sLst, oLst = zip(*multiLst)
    zLst = range(len(oLst))
    multiLst = list(zip(oLst, zLst))
    multiLst.sort()
    oLst, zLst = zip(*multiLst)

if uvCovArr is not None:
    for i, e in enumerate(uvCovArr):
        u = e["uArr_lam"]
        v = e["vArr_lam"]
        
        plt.scatter(x=u/1000, y=v/1000, marker=".", edgecolors='none',
                    s=2, color=colLst[i%len(colLst)], zorder=zLst[i])
        plt.scatter(x=u/1000, y=-v/1000, marker=".", edgecolors='none',
                    s=2, color=colLst[i%len(colLst)], zorder=zLst[i])
plt.show()
```

# Load the Model and Set the Pixel Scale in Arcsec

```python
obsMan.load_model_image("models/galaxy_lobes.png")
obsMan.set_pixscale(1.0)
```

# Calculate the FFT of the Model Image

```python
obsMan.invert_model()
```

## To Plot The FFT

```python
fftArr = obsMan.modelFFTarr
```

```python
paramDict = obsMan.get_scales()
limit = paramDict["fftScale_lam"]/1e3
```

```python
extent = [-limit, limit, -limit, limit]
plt.imshow(np.abs(fftArr), norm=LogNorm(), cmap=plt.cm.cubehelix,
          interpolation="nearest", origin="lower", extent=extent)

plt.show()
```

# Grid the UV-Coverage onto the Same Pixels as the FFT as the Model Image

```python
obsMan.grid_uvcoverage()
```

# Create the beam image

```python
obsMan.calc_beam()
```

## Plot Synthesised Beam

```python
beamArr = np.abs(obsMan.beamArr)
pRange = (-0.1, 0.5)
```

```python
zMin = np.nanmin(beamArr)
zMax = np.nanmax(beamArr)
zRng = zMin - zMax
zMin -= zRng * pRange[0]
zMax += zRng * pRange[1]
```

```python
plt.imshow(beamArr, cmap=plt.cm.cubehelix, interpolation="nearest",
          origin="lower", vmin=zMin, vmax=zMax)
plt.show()
```
